### To Know sub packages in packages (libraries)

- math 
- OpenCV
- Numpy
- matplotlib

- If we want we can check for other version and their sub packages like this


```python
import math

for i in dir(math):
    print('math.',end="")
    print(i)
```

```python
import cv2
print("OpenCV version : ",cv2.__version__,"\n\n")
for i in dir(cv2):
    print('cv2.',end="")
    print(i)
```

```python
import matplotlib
print("Matplotlib version : ",matplotlib.__version__,"\n\n")
for i in dir(matplotlib):
    print('matplotlib.',end="")
    print(i)
```

```python
import numpy
print("numpy version : ",numpy.__version__,"\n\n")
for i in dir(numpy):
    print('numpy.',end="")
    print(i)
```
