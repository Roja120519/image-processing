### Drawing
- Text
- Line
- Circle
- Square
- Rectangle
- Ellipse
- Polygon

- We can draw these in an image or in an empty image window which we can take by using numpy
- input
    - image
    - color
    - thickness
    - line type
    


```python
import numpy as np
import cv2
img1=np.zeros((900,900,3),np.uint8)
#img1=cv2.imread("standard_test_images\\standard_test_images\\1.tiff")
c=1
while c==1:
    print("1.Line\n2.rectangle\n3.circle\n4.ellipse\n5.polygone\n6.text\n")
    value=int(input())
    if value==1:
        sx,sy=map(int,input("Enter starting point of line values x y  ").split())
        ex,ey=map(int,input("Enter ending point of line values x y  ").split())
        t=int(input("Enter thickness value  "))
        cv2.line(img1, (sx,sy), (ex,ey), (255,0,0), t)
    elif value==2:
        sx,sy=map(int,input("Enter starting point rectangle values x y  ").split())
        ex,ey=map(int,input("Enter ending point rectangle values x y  ").split())
        t=int(input("Enter thickness value  "))
        cv2.rectangle(img1, (sx,sy), (ex,ey), (0,255,0), t)  
    elif value==3:
        cx,cy=map(int,input("Enter center point values x y  ").split())
        t=int(input("Enter thickness value  "))
        radius=int(input("Enter radius value  "))
        cv2.circle(img1, (cx,cy), radius, (0,0,255), t)   
      
    elif value==4:    
        #cv2.ellipse(image, (major_axes, minor_axes), rotation_degree, remove_some_part, how_many_degrees_the_shape_comes, color, thick_ness )
        major_x,major_y=map(int,input("Enter major axes point values").split())
        minor_x,minor_y=map(int,input("Enter minor axes point values").split())
        t=int(input("Enter thickness value  "))
        r=int(input("Enter rotation degrees  "))
        Remove_some_part=int(input("Enter how much part u want to remove "))
        shape_degree=int(input("Enter how many degrees the shape comes"))
        cv2.ellipse(img1, (major_x,major_y),(minor_x,minor_y), r,Remove_some_part,shape_degree, (127,120,255), t)

    elif value==5:
        #POLYGONE
        points=np.array([[80,2], [125,10], [170,10], [230,5], [30,50]],np.int32)
        points=points.reshape(-1,1,2)
        cv2.polylines(img1, [points], True, (0,255,255),2)   
    elif value==6:
        text1=input("Enter the text which u want to place in image  ")
        sx,sy=map(int,input("Enter startingmpoint values  ").split())
        cv2.putText(img1, text1, (sx,sy),cv2.FONT_HERSHEY_SIMPLEX, 2, (255,225,0))

    cv2.imshow('img',img1)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
    print("\nEnter 1 to continue\nEnter 0 to exit\n")
    c=int(input())
cv2.imwrite("standard_test_images\\output\\drawings.tiff",img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
    
```
