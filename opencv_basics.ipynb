###  OpenCV basics
- reading an image from particular path   
- reading images in different modes
    - we have 3 types of modes
        1. 1 --> Loads a color image. Any transparency of image will be neglected. It is the default flag.
        2. 0 --> To read gray scale image
        3. -1 --> To read unchanged images (Loads image as such including alpha channel)
            
- displaying image window
- displaying multiple images

- Write an image (Save an image in particular folder)

- Display multiple images using matplotlib.pyplot library 




```python
# To read 3 modes of images and displaying.displaying multiple images in differend windows

import cv2

gray=cv2.imread("standard_test_images\\standard_test_images\\1.tiff",0)
color=cv2.imread("standard_test_images\\standard_test_images\\1.tiff",1)
unchange=cv2.imread("standard_test_images\\standard_test_images\\1.tiff",-1)

#color=cv2.IMREAD_COLOR("C:\\Users\\ROJA MOGILI\\Downloads\\standard_test_images\\standard_test_images\\1.tiff")
#gray=cv2.IMREAD_GRAYSCALE("C:\\Users\\ROJA MOGILI\\Downloads\\standard_test_images\\standard_test_images\\1.tiff")
#unchange=cv2.IMREAD_UNCHANGED("C:\\Users\\ROJA MOGILI\\Downloads\\standard_test_images\\standard_test_images\\1.tiff")

names=[gray,color,unchange]
windows=["gray","color","unchange"]

for i in range(0,len(names),1):
    cv2.imshow(windows[i],names[i])
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# To write an image in particular path
import cv2
img=cv2.imread("standard_test_images\\standard_test_images\\1.tiff",1)
print("\n1.Save image\n2.Display image")
value=int(input())
if value==1:
    cv2.imwrite("standard_test_images\\output\\write_img.tiff",img)
elif value==2:
    cv2.imshow("img",img)
    k=cv2.waitKey(0) & 0xFF
    if k==27:
        cv2.destroyAllWindows()
```

```python
# Display multiple images using matplotlib.pyplot library

# It will display in different colors because matplotlib an openCV are different in color images 
#  OpenCv images in RGB formate
#  PMAtplotlib images in BGR formate

import cv2
import matplotlib.pyplot as plt

gray=cv2.imread("standard_test_images\\standard_test_images\\1.tiff",0)
color=cv2.imread("standard_test_images\\standard_test_images\\1.tiff",1)
unchange=cv2.imread("standard_test_images\\standard_test_images\\1.tiff",-1)

names=[gray,color,unchange]
windows=["gray","color","unchange"]

for i in range(0,len(names),1):
    plt.subplot(2,4,i+1)
    plt.imshow(names[i])
    plt.title(windows[i])
    plt.xticks([])
    plt.yticks([])
    
```
