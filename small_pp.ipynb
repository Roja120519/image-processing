```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('standard_test_images\\standard_test_images\\1.tif')
rows,cols,channles=img.shape
rows,cols
```

```python
parts={}
c=0
r=0
p=1
devide=rows//2
while c<cols and r<rows:
    if p%4==1 or p%4==0:
        parts[p]=img[r:r+devide,c:c+devide]
        p=p+1
        print(r,c)
    
    elif p%4==2 or p%4==3:
        r=r+devide
        parts[p]=img[r:r+devide,c:c+devide]
        print(r,c)
        p=p+1
        parts[p]=img[c:c+devide,r:r+devide]
        p=p+1
        c=c+devide
        print(r,c)
        
    

for key,value in parts.items():
    print(key)
    
    
cv2.imshow("part1",parts[1])
plt.subplot(251),plt.imshow(parts[2])
plt.subplot(252),plt.imshow(parts[1])
plt.subplot(253),plt.imshow(parts[3])
plt.subplot(254),plt.imshow(parts[4])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
